```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#  Question 1

x=np.arange(40,50)
y=np.arange(50,60)
plt.plot(x,y,'ro--',linewidth=5,markersize=10)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#Question 2

days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
plt.figure(figsize=(5,5), dpi=100)
plt.plot(days, sales_1)
plt.plot(days, sales_2)

plt.xlabel('days of the week')
plt.ylabel('sales')
plt.grid()
plt.show()
```
